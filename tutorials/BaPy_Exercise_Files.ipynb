## **File I/O**

#### **Python File Operation**

In Python, file operations follow a certain order:


1. Opening a file

2. Reading or writing

3. Closing a file


#### **Opening Files**

The method `open() `is used to open files.

We already have a file called `my_file.txt`.

Python opens files in the read mode by default.

```python
my_file = open("my_file.txt")

print(my_file)
```

#### **Reading Files**

When we open a file by using `open() `method, now we can read its content with `read() `method.

```python
read_my_file = my_file.read()

print(read_my_file)
```

#### **Closing Files**

`close()` method allows you to close a file, which in turn will free up the resources that are linked to the file.

```python
my_file.close()
```

#### **Creating a File**

```python
file2 = open("my_file2.txt", "x")

file2.close()
```

**What is the difference between creating a file by `x` and `w`?**

- `x` mode creates a new file exclusively and ensures that the file does not already exist.

- `w`, on the other hand, creates a new file or overwrites an existing one. It also erases its previous contents.

#### **Writing a File**

```python
file2 = open("my_file2.txt", "w")

file2.write("I need an ice cream asap.")

file2.close()
```

#### **Using `with` Syntax**

`with`syntax allows you to automatically close files as soon as you are at the end of the `with ` block.

It provides more readable code structure than the manual opening and closing of files.

```python
#Open the file using with statement
with open("my_file3.txt", "r") as file:
    #Read the contents of the file
    file_contents = file.read()
    print(file_contents)

#File is automatically closed outside the "with" block
```

We can write the content of.a file to another file.

```python
#Open the output file using with statement
with open("output_file.txt", "w") as output_file:
    #Write the contents to the output file
    output_file.write(file_contents)
    print(output_file)
```

We can print each character of a file using a for loop and iterate over the file object as follows:

```python
#Open the file
with open("output_file.txt", "r") as file:
    #Iterate over each character in the file
    for character in file.read():
        print(character)
```

#### **File Navigation and Operations:**

1.  Seeking a Specific Position in a File:

> * When working with files, it's important to understand the concept of the file position, which represents the current location in the file.
* We can use the `seek() `method to move the file position to a specific byte or character.




```python
#Open the file in write mode
file = open('example.txt', 'w')

#Write content to the file
file.write('This is a test file.\n')
file.write('It contains some text.\n')

#Move the file position to the 10th byte
file_search = file.seek(10)
print(file_search)

#Close the file
file.close()

print("File created and modified successfully.")
```

2. Getting the Current File Position:


> * Another useful function is `tell()`, which allows us to get the current file position.
* It can be handy for keeping track of the file position during file operations.


```python
file = open('example.txt', 'r')

#Get the current file position
position = file.tell()
print("Current position:", position)

#Close the file
file.close()
```

#### **Checking File Existence and Properties**

It's often useful to check if a file exists or to obtain its properties. We can use the os.path module for these tasks.

```python
import os.path

#Check if a file exists
exists = os.path.exists("example.txt")
print(exists)

#Check if it's a regular file
is_file = os.path.isfile("example.txt")
print(is_file)

#Get the file size
size = os.path.getsize("example.txt")
print(size)
```

#### **File Error Handling**

Error handling is crucial when working with files. We need to handle exceptions such as file not found or permission errors. We handle these exceptions using try-except blocks.:



```python
try:
    file = open("output_file1.txt", "r")
    #Perform file operations
except FileNotFoundError:
    print("File not found!")
except PermissionError:
    print("Permission denied!")
```

#### **Working with CSV Files**

CSV (Comma-Separated Values) files are widely used for storing tabular data. They consist of rows, each containing data separated by commas. Python provides the csv module to handle CSV files efficiently. We can use the csv module's `reader()` function to read CSV data into lists or DictReader class to read data into dictionaries. Similarly, we can use the `writer()` function or DictWriter class to write data to CSV files.
The csv module offers various options for customizing the behavior, such as specifying delimiters, handling headers, and handling missing data.



```python
import csv

with open("my_data.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
