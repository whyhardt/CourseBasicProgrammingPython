### **Creating a Simple Line Plot**

We can create a simple line plot by using `plt.plot(x, y) ` method.

Here `x` and `y `correspond to the data you have.

### **Customizing Plots**


* Customizing plots by adding labels, titles, legends, and gridlines.

* Modifying colors, line styles, marker styles, and other visual elements to enhance the aesthetics of the plot.

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Defining data, line color, line style, marker symbol and label.
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Data')

# Labelling the x-axis.
plt.xlabel('X-axis')

# Labelling the y-axis.
plt.ylabel('Y-axis')

# Adding a title to the plot.
plt.title('Simple Line Plot')

# Adding grid lines to the plot.
plt.grid()

# Displaying the label we have set in the legend.
plt.legend()

plt.show()
````

### **Creating a Histogram**

The `plt.hist(data)` method allows us to create histograms.

````python
data = [1, 1, 1, 2, 2, 3, 3, 4, 4, 5]
plt.hist(data)
plt.show()
````

It is possible to adjust both the edge and bar colors.

````python
plt.hist(data, color='green', edgecolor='black')
plt.show()
````

The`bins` parameter is used to control the intervals which allows you to visualize the data in different levels of resolution.

````python
plt.hist(data, bins=18, color='green', edgecolor='black')
plt.show()
````

We can set custom ticks for both axes by using `plt.xticks ` and `plt.yticks`.

````python
plt.hist(data, bins=18, color='green', edgecolor='black')

# Setting custom x-axis tick positions.
plt.xticks([1, 2, 3, 4, 5])

# Setting custom y-axis tick positions.
plt.yticks([1, 2, 3])

plt.show()
````

## **More on NumPy**

### **Iterating over Array Elements**

Involves traversing or accessing each element in an array one by one. Iteration allows you to perform operations on individual elements or apply functions to each element.

### **Iterating with For Loop**

 Inside the loop, we access each element using the iterator variable. By iterating over the elements of a 1-dimensional array using a for loop, you can access, manipulate, or perform calculations on each element individually. This gives you flexibility and control over the elements of the array during the iteration process.

````python
import numpy as np

# Create a 1-dimensional array
arr1 = np.array([1, 2, 3, 4, 5])

# Iterate over the elements of the array using a for loop
for element in arr1:
    # Access and print each element
    print(element)
````

### **Iterating with `nditer`**

A powerful function in NumPy that allows flexible and efficient iteration over multi-dimensional arrays. With nditer, we can access and manipulate each element of an array without the need for explicit loops.
The basic syntax of nditer is as follows:

```
numpy.nditer(array, flags, op_flags, op_dtypes)
```
The array parameter represents the input array over which iteration will be performed. flags, op_flags, and op_dtypes are optional parameters that control the iteration behavior.
* Flags: The flags parameter allows us to specify various options for iteration. Here are some commonly used flag values:

> 1. 'readwrite': Allows both reading and writing of array elements.
2. 'readonly': Restricts access to read-only mode.
3. 'writeonly': Restricts access to write-only mode.
4. 'external_loop': Efficient iteration when the array has multiple dimensions.

* Op_flags:The op_flags parameter controls specific iteration behavior. Some useful values include:

> 1. 'reduce': Performs reduction operations on arrays, such as sum, product, etc.
2. 'no_broadcast': Prevents broadcasting of dimensions during iteration.
3. 'ignore_broadcast_dimensions': Ignores dimensions that are broadcasted during iteration.

* Order: The order parameter specifies the order in which elements are accessed during iteration. Common values are:

>1. 'C': Row-major order, iterating over the elements in each row first.
2. 'F': Column-major order, iterating over the elements in each column first.
3. 'A': Accesses elements in a memory-efficient way.
4. 'K': Accesses elements in the order specified by the strides.

````python
import numpy as np

# Create a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

# Iterate over the elements using nditer
for element in np.nditer(arr2):
    print(element, end=" ")
````

````python
# Create a 3D array
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
#print(arr3)

# Iterate over the elements in a specific order
for element in np.nditer(arr3, order='F'):
    print(element, end=" ")
````

### **Iterating over Multiple Arrays Using nditer**

````python
arr4 = np.array([1, 2, 3])
arr5 = np.array([4, 5, 6])

with np.nditer([arr4, arr5]) as it:
    for x, y in it:
        # Perform operations or calculations on x and y
        # Print or store the results
        print(x, y)
````

nditer treats the input arrays as read-only by default. If modification of the array elements is required, you can pass the op_flags parameter to specify write access.

````python
with np.nditer([arr4, arr5], op_flags=['readwrite']) as it:
    for x, y in it:
        # Perform operations or calculations on x and y
        # Modify the elements if needed
        x += 1
        y *= 2
        print(x, y)
````
